### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
```

### Step 4. Which continent drinks more beer on average?

```python

beer_continent = drinks.groupby('continent').beer_servings.mean().sort_values(ascending=False)
print("Continent with max beer average:")
print(beer_continent.head(1))
```

### Step 5. For each continent print the statistics for wine consumption.

```python

wine_stats = drinks.groupby('continent').wine_servings.describe()
print("\nWine consumption statistics per continent:")
print(wine_stats)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python

mean_all = drinks.groupby('continent').mean(numeric_only=True)
print("\nMean consumption per continent (all columns):")
print(mean_all)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python

median_all = drinks.groupby('continent').median(numeric_only=True)
print("\nMedian consumption per continent (all columns):")
print(median_all)

```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python

spirit_stats = drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
print("\nSpirit consumption (Mean, Min, Max):")
print(spirit_stats)
```
